string_frac(in_string): This function converts a string input (including fractions) to a float. It checks if the input string contains a "/", indicating a fraction. If it does, it splits the string at the "/" and converts the numerator and denominator to floats, calculates the division, and returns the result. If there is no "/", it simply converts the input string to a float and returns it.

one_step_add(): This function generates a simple addition problem by randomly choosing two numbers, "a" and "b." It displays the equation "x + a = b" to the user and prompts them to input the value of "x." It calculates the correct answer by subtracting "a" from "b." If the user's input matches the correct answer, it prints "Correct!" Otherwise, it displays "Try again" and shows the correct answer.

one_step_subtract(): Similar to one_step_add(), this function generates a subtraction problem by randomly selecting two numbers, "a" and "b." It displays the equation "a + x = b" to the user and prompts them to input the value of "x." It calculates the correct answer by subtracting "a" from "b." If the user's input matches the correct answer, it prints "Correct!" Otherwise, it displays "Try again" and shows the correct answer.

one_step_mult(): This function generates a multiplication problem by randomly choosing two numbers, "a" and "b." It displays the equation "a x x = b" to the user and prompts them to input the value of "x." It calculates the correct answer by dividing "b" by "a." It uses the string_frac() function to handle potential fractional inputs from the user. If the user's input, after conversion to a float, matches the correct answer, it prints "Correct!" Otherwise, it displays "Try again" and shows the correct answer.


```python
# converts string input (including fractions) to float

def string_frac(in_string):
  if"/" in in_string:
    nd = in_string.split("/")
    n = float(nd[0])
    d = float(nd[1])
    ans = n/d
    return ans
  else:
    ans = float(in_string)
    return ans
```

```python
# Simplest one-step addition
def one_step_add():
    import random
    # Display problem
    a = random.randint(-4,10)
    b = random.randint(2,24)
    print("x + ", a, " = ", b)
    ans = float(input("x = "))
    answer = b-a
    # Test input
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
```

```python
# One-step additon with negaive numbers
def one_step_subtract():
    import random
    a = random.randint(-19,-1)
    b = random.randint(2,24)
    print(a, " + x = ", b)
    ans = float(input("x = "))
    # test
    answer = b-a
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
```

```python
# One-step multiply
def one_step_mult():
    # Uses string_frac(<input string>)
    import random
    a = random.randint(1,11)
    b = random.randint(2,24)
    print(a, "x = ", b)
    print("Round your answer to two decimal places.")
    ans_in = (input("x = "))
    answer = round(b/a,2)
    # test
    if string_frac(ans_in)==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
```

```python

# One-step divide
def one_step_div():
    import random
    a = random.randint(1,11)
    b = random.randint(2,24)
    print("x/", a, " = ", b)
    ans = float(input("x = "))
    answer = b*a
    # test
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
```

```python

# Two-step problems
def two_step():
    import random
    # Uses string_frac()
    a = random.randint(1,11)
    b = random.randint(-7,12)
    c = random.randint(2,36)
    print(a, "x + ", b, " = ", c)
    print("Round answer to two decimal places")
    ans_in = input("x = ")
    answer = (c-b)/a
    # test
    if round(string_frac(ans_in),2)==round(answer,2):
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")


```

```python
# get string input, which will include a decimal point
digits = input("enter a decimal number to convert:")

# get number of decimal places as an integer
exponent = int(len(digits))-1

#conert the input to a float number
n = float(digits)

#use the exponent to get the numerator
numerator = int(n*10**exponent)

# #use the exponent to get the denominator
denominator = 10**exponent

# percent is the first two decimal numbers
percent = n*100

#output
print("the decimal is ", n)
print("the fraction is ", numerator, "/", denominator)
print("the percent is ", percent, "%")
```
